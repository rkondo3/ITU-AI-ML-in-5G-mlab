```python
import pickle
import pandas as pd

import glob
import os

vi_dir = "/data/output-directory_all/df/physical-infrastructure-bgpnw2"
# vi_dir = "/data/output-directory_all/df/virtual-infrastructure-bgpnw2/networks/"
# nd_dir = "/data/output-directory_all/df/network-device-bgpnw2"

pkl_list = glob.glob(os.path.join(vi_dir, "*.pkl"))#get file path list

for pkl_path in pkl_list:
    df_from_pkl = pd.read_pickle(pkl_path)
    df_from_pkl=df_from_pkl.replace(np.nan, -9999)
    df_drop = pd.DataFrame(data=None, index=None, columns=None, dtype=None, copy=False)
    
    for i in range(len(df_from_pkl.columns)):#
        s = {None}
        row = df_from_pkl.iloc[:,i]

        for j in range(len(row)):#列内の要素数を数える
            temp_value = row.iat[j]
            if temp_value not in s:
                s.add(str(temp_value))

        if len(s) > 2:
            #print("hoge")
            print(df_from_pkl.columns[i])
            print(s)
            df_drop[df_from_pkl.columns[i]]=row #列を追加する
            
    filename = os.path.splitext(os.path.basename(pkl_path))[0] + "_dropped.pkl"
    #filename = os.path.basename(pkl_dir)
    save_dir = "/data/output-directory_all/df/physical-infrastructure-bgpnw2/dropped_df2/"
    write_path = os.path.join(save_dir, filename)

    df_drop.to_pickle(write_path)
```
