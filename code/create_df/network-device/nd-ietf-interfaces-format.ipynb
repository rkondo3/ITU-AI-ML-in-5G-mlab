```python
def get_IntGW(nd_dic_device):
    if 'ietf-interfaces' not in nd_dic_device['modules']:
        return {}
    ietf_interfaces_dic = nd_dic_device['modules']['ietf-interfaces']
    ret_dic = {}
    for i in range(len(ietf_interfaces_dic['interfaces']['interface'])):
        interface = ietf_interfaces_dic['interfaces']['interface'][i]
        if (interface['name'] == 'GigabitEthernet1'):
            ret_dic['GigabitEthernet1_type'] = interface["type"]
            ret_dic['GigabitEthernet1_enabled'] = interface["enabled"]
            ret_dic['GigabitEthernet1_description'] = interface["description"]
        elif (interface['name'] == 'GigabitEthernet2'):
            ret_dic['GigabitEthernet2_type'] = interface["type"]
            ret_dic['GigabitEthernet2_enabled'] = interface["enabled"]
            ret_dic['GigabitEthernet2_description'] = interface["description"]
        elif (interface['name'] == 'GigabitEthernet3'):
            ret_dic['GigabitEthernet3_type'] = interface["type"]
            ret_dic['GigabitEthernet3_enabled'] = interface["enabled"]
            ret_dic['GigabitEthernet3_description'] = interface["description"]
        elif (interface['name'] == 'GigabitEthernet4'):
            ret_dic['GigabitEthernet4_type'] = interface["type"]
            ret_dic['GigabitEthernet4_enabled'] = interface["enabled"]
            ret_dic['GigabitEthernet4_description'] = interface["description"]
        elif (interface['name'] == 'GigabitEthernet5'):
            ret_dic['GigabitEthernet5_type'] = interface["type"]
            ret_dic['GigabitEthernet5_enabled'] = interface["enabled"]
            ret_dic['GigabitEthernet5_description'] = interface["description"]
        elif (interface['name'] == 'GigabitEthernet6'):
            ret_dic['GigabitEthernet6_type'] = interface["type"]
            ret_dic['GigabitEthernet6_enabled'] = interface["enabled"]
            ret_dic['GigabitEthernet6_description'] = interface["description"]
        elif (interface['name'] == 'GigabitEthernet7'):
            ret_dic['GigabitEthernet7_type'] = interface["type"]
            ret_dic['GigabitEthernet7_enabled'] = interface["enabled"]
            ret_dic['GigabitEthernet7_description'] = interface["description"]
        elif (interface['name'] == 'Loopback1'):
            ret_dic['Loopback1_type'] = interface["type"]
            ret_dic['Loopback1_enabled'] = interface["enabled"]
        elif (interface['name'] == 'VirtualPortGroup0'):
            ret_dic['VirtualPortGroup0_type'] = interface["type"]
            ret_dic['VirtualPortGroup0_enabled'] = interface["enabled"]
        else:
            print("%s is not defined" % interface['name'] )
            
    return ret_dic



print(get_IntGW(vi_dic_list[6]['devices'][1]))
```
